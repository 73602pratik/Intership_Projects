# Handling Outliers

```python
from scipy import stats
```

```python
## using IQR to remove the outliers LotFrontage
IQR_LotFrontage = stats.iqr(data.LotFrontage, interpolation = 'midpoint')

Q1_LotFrontage = data.LotFrontage.quantile(0.25)

Q3_LotFrontage = data.LotFrontage.quantile(0.75)

min_limit_LotFrontage = Q1_LotFrontage - 1.5*IQR_LotFrontage

max_limit_LotFrontage = Q3_LotFrontage + 1.5*IQR_LotFrontage

data.loc[(data.LotFrontage > max_limit_LotFrontage) | (data.LotFrontage < min_limit_LotFrontage), 'LotFrontage']=np.median(data.LotFrontage)
```

```python
## using IQR to remove the outliers LotArea
IQR_LotArea = stats.iqr(data.LotArea, interpolation = 'midpoint')

Q1_LotArea = data.LotArea.quantile(0.25)

Q3_LotArea = data.LotArea.quantile(0.75)

min_limit_LotArea = Q1_LotArea - 1.5*IQR_LotArea

max_limit_LotArea = Q3_LotArea + 1.5*IQR_LotArea

data.loc[(data.LotArea > max_limit_LotArea) | (data.LotArea < min_limit_LotArea), 'LotArea']=np.median(data.LotArea)
```

```python
## using IQR to remove the outliers Firstflrsf
IQR_Firstflrsf = stats.iqr(data.Firstflrsf, interpolation = 'midpoint')

Q1_Firstflrsf = data.Firstflrsf.quantile(0.25)

Q3_Firstflrsf = data.Firstflrsf.quantile(0.75)

min_limit_Firstflrsf = Q1_Firstflrsf - 1.5*IQR_Firstflrsf

max_limit_Firstflrsf = Q3_Firstflrsf + 1.5*IQR_Firstflrsf

data.loc[(data.Firstflrsf > max_limit_Firstflrsf) | (data.Firstflrsf < min_limit_Firstflrsf), 'Firstflrsf']=np.median(data.Firstflrsf)
```

```python
## using IQR to remove the outliers GrLivArea
IQR_GrLivArea = stats.iqr(data.GrLivArea, interpolation = 'midpoint')

Q1_GrLivArea = data.GrLivArea.quantile(0.25)

Q3_GrLivArea = data.GrLivArea.quantile(0.75)

min_limit_GrLivArea = Q1_GrLivArea - 1.5*IQR_GrLivArea

max_limit_GrLivArea = Q3_GrLivArea + 1.5*IQR_GrLivArea

data.loc[(data.GrLivArea > max_limit_GrLivArea) | (data.GrLivArea < min_limit_GrLivArea), 'GrLivArea']=np.median(data.GrLivArea)
```

# missing values

```python
#  Now lets check for numerical variables the contains missing values
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']
# we will print the numercial nan variables and percentage of missing values
for feature in features_nan:
    print('{}: {}% missing value'.format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
# replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data[features_nan]=data[features_nan].fillna('missing')
    return data
dataset=replace_cat_feature(dataset,features_nan)
```

```python
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# now for numerical_feature variables the contains missing values
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

for feature in numerical_with_nan:
    print('{}: {}% missing value'.format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
# replace missing numerical values
for feature in numerical_with_nan:
    # we will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    # create new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
```

# Handling Rare Catergorical Feature

### We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
len(categorical_features)
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(50)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k,in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
models = {
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "XGBRegressor": XGBRegressor(), 
    "AdaBoost Regressor": AdaBoostRegressor(),
    'GradientBoostingRegressor': GradientBoostingRegressor(),
    'SVR': SVR()
}
model_list = []
r2_list =[]

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train, y_train) # Train model

    # Make predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    # Evaluate Train and Test dataset
    model_train_mae , model_train_rmse, model_train_r2 = evaluate_model(y_train, y_train_pred)

    model_test_mae , model_test_rmse, model_test_r2 = evaluate_model(y_test, y_test_pred)

    
    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
    print("- R2 Score: {:.4f}".format(model_train_r2))

    print('----------------------------------')
    
    print('Model performance for Test set')
    print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
    print("- R2 Score: {:.4f}".format(model_test_r2))
    r2_list.append(model_test_r2)
    
    print('='*35)
    print('\n')
```
